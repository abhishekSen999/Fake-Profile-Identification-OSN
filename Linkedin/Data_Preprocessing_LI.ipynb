```python
import csv
import pandas as pd
import numpy as np
```

```python
fake_user_data = pd.read_csv('data/fake_user.csv')
genuine_user_data = pd.read_csv('data/genuine_user.csv')
```

```python
features = ['connectionsCount',
             'experienceCount',
             'educationCount',
             'volunteerCount',
             'skillsCount',
             'recommendationsCount',
             'honorsCount',
             'courseCount',
             'projectsCount',
             'testCount',
             'publicationCount',
             'bgImage']
features
```

```python
genuine_user_data = genuine_user_data[features]
genuine_user_data.head()
```

```python
fake_user_data = fake_user_data[features]
fake_user_data.head()
```

```python
#Labelling Genuine and Fake users accordingly. 
'''
    label  0  = Genuine users
    label  1  = Fake users
''' 
genuine_user_data['label'] = 0
fake_user_data['label'] = 1

total_features = genuine_user_data.shape[1]
total_features
```

```python
#Combining the labelled Genuine and Fake users datasets. 

dataset = pd.concat([genuine_user_data, fake_user_data], axis = 0, sort = 'False')
dataset.head()
```

```python
dataset['bgImage'].replace(('YES', 'NO'), (1, 0), inplace=True)
```

```python
dataset['bgImage'] = pd.to_numeric(dataset['bgImage'])
```

```python
# Re-order Columns
features = ['connectionsCount','experienceCount','educationCount','volunteerCount','recommendationsCount','honorsCount','courseCount','projectsCount','testCount','publicationCount','bgImage','skillsCount','label']
dataset = pd.DataFrame(dataset, columns = features)
dataset.head()
```

```python
export_csv = dataset.to_csv ('data/LI_dataset.csv', index = None, header=True) 
```
